```python
import itertools as it
import pandas as pd

"""
coding parking functions
"""


def parking_experiment(x, y):
    x = tuple(x_i - 1 for x_i in x)
    defect = 0
    n = len(y)
    m = sum(y)
    spots = [None for _ in range(m)]
    for j in range(n):
        parked = False
        for i in range(x[j], m - y[j] + 1):
            if all(spots[i + k] is None for k in range(y[j])):
                for k in range(y[j]):
                    spots[i + k] = j
                parked = True
                break
        if not parked:
            defect += 1

    return defect
```

```python
"""
This part of the code verifies the sufficient and necessary conditions for a 3-tuple car length vector to be minimally
invariant. It verifies all iterations that obeys the negation the conditions. If there is at least one case that fails, it
will print out where the failure occurs. The failure occures if there is a case in this codes that disagrees with the parking
experiment.
""" 

# --- m.i. for n = 3 --- #
k = 5
n = 3

print('Processing n: {0}'.format(n))
ct = 0
tot = k ** n
for y in it.product([i for i in range(1, k + 1)], repeat=n):
    ct += 1
    print('     {0}%'.format(round(100 * ct/tot, 2)), sep=' ', end='\r', flush=True)
    m = sum(y)

    mi_theorem = True
    for w in range(2, m + 1):
        x = tuple([1 for _ in range(n - 1)] + [w])
        if all(parking_experiment(perm_x, y) == 0 for perm_x in it.permutations(x)):
            mi_theorem = False
            break

    mi_w_condition = True
    for w in range(2, m + 1):
        if (
                (w <= 1 + y[0] + y[1]) and
                ((w <= 1 + y[0]) or (w == 1 + y[0] + y[2])) and
                ((w == 1 + y[1]) or (w == 1 + y[1] + y[2]) or ((y[1] >= y[2]) and (w == 1 + y[2])))
        ):
            mi_w_condition = False

    mi_condition = False
    if not ((y[1] <= y[0]) or ((y[1] == y[0] + y[2]) and (y[2] <= y[1])) or ((y[1] + y[2] <= y[0]) and (y[2] <= y[0])) or (
                (y[1] == y[0]) and (y[2] <= y[0])) or ((y[2] <= y[0]) and (y[2] <= y[1]))):
        mi_condition = True

    mi_ineq_condition = True
    if not ((y[0] < y[1]) and (y[0] < y[2]) and (y[1] != y[0] + y[2])):
        mi_ineq_condition = False

    if mi_theorem != mi_w_condition or mi_w_condition != mi_condition or mi_condition != mi_ineq_condition:
        input('     Failure: {0}'.format(y))
```
